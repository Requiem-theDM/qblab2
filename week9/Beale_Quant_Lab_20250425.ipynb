**KEY DETAIL**

When you perform a Fourier Transform, you actually generate a solution that is COMPLEX, i.e. the values (space or time) have real and imaginary parts:

value = x +iy

Where "i" is $\sqrt{-1}$. 

What you plotted before were the absolute values of this, which essentially collapsed the imaginary contribution.

HOWEVER, when you took the INVERSE Fourier transform, you used the complex numbers, because the imaginary part of the Fourier solution tells you the PHASE of the waves. This is very important if you want to reconstruct the original image because the phase tells you the relative location of the features in the real space.

So, why am I telling you this?

When you are given a diffraction pattern (like an X-ray diffraction pattern), all you are given is the REAL information, i.e. the intensity of the diffraction patterns. You do not have any phase information, which means YOU CANNOT reconstruct the original structure!!!

This is THE CENTRAL challenge in inferring structure from X-ray diffractions. There are a variety of tricks that can be performed to determine the phase, but we won't go over that here.

To make things easier for you, the diffraction pattern I have provided DOES have the phase information.

i.e. the numpy array you will download has COMPLEX values. This will allow you to reconsctruct the original structure.

```python
def ift_2d(ft : np.typing.ArrayLike) -> np.typing.NDArray:
    """Takes the Fourier transform of an image as input and returns the real component of the inverse transform."""
    return np.fft.fftshift(np.fft.ifft2(np.fft.ifftshift(ft))).real
```
